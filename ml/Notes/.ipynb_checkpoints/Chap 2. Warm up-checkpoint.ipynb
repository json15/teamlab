## Understanding of Data

#### 모델을 학습할때 영향을 주는 것들
y = ax + b
y = 종속변수
a, b = 파라미터 (알고리즘을 통해 최적값을 찾아야한다)
x = 독립변수

#### What is Feature?
- ML에서 data의 특징을 나타내는 변수
- Feature = 독립변수 = input 변수
- Table 상의 data에서 coloumn을 의미한다
- 하나의 data instance(실제 데이터)는 feature vecore로 표기한다

#### Feature vector 표기법
$x_j^(i)$
- j는 몇번째 column의 element 인지 표기
- i는 몇번째 row의 element 인지 표기

#### 공식
$ y = w_1x_1 + w_2x_2 + w_nx_n +w_0x_0 = \sum_{j=0} w_jx_j = w^Tx$
- $w_0x_0$은 상수항이다

#### Feture 가 n 개일때
- n차원 n dimension으로 차원이 늘어나고, 이렇게 차원이 늘어나는 것을 차원의 저주(curse of dimensionality) 라고 한다

#### 차원의 저주
- Data의 차원이 증가할수록 데이터를 표현하는 공간이 증가한다
    1. 희박한 벡터가 증가한다(값이 없는 feture가 늘어난다)
    2. 샘플데이터가 늘어난다(data의 분포, 모델 추정에 어려운이 생긴다)
    

## Data Types

Feature 별로 데이터의 유형이 다르다.

#### DB의 두가지 종류
1. 연속형 값 continuous
    - 값이 끊어지지 않고 연결됨
    ex) 온도, 평균 점수, 속도
    - 일반적인 실수의 값들이다
2. 이산형값 discrete
    - 값이 연속적이지 않은 값
    ex) 성별, 우편주소, 등수 등
    - labbel로 이루어지는 값들

#### 다른방식으로 나누어 보자
1. 숫자형값 Numeric Type
    - 정량적으로 측정 가능한 data type(실수, 정수)
    - 단위(Scale)이 있는 Inerval-scaled type
    - 비율이 있는 ratio - scaled type
    @@ 단위대로 증가해야한다
    즉, 1,2,3,4가 있다고 한다면 2가 1의 2개여야하고 계속적으로 해당 단위로 나가야한다
2. Nominal Type(명목척도)
    - Category(범주)로 분류가 가능한 Type
    - 색, 학교명, ID
    - 2개의 category로 이루어져있을 경우 binary type이라고 부른다
3. ordinal type
    - 범주로 분류가 가능하나 범주한에 순서가 있는 data
    ex) 5점척도, 병크기(대, 중, 소)
    - Why not numeric type?
        - 사람마다 scale, unit 이 다르기 때문이다
        - 순서가 있는 것 =! 배수로 증가하는것
        
#### Feture 별로 달라서 생길수 있는 문제들
- 하나의 data table도 feature 별로 Type이 다를수 있기 때문에 처리를 해야한다
- 또, 값이 크기가 크고 작은 애들이 있기 때문에 weight로 맞춰줘야 하는 작업도 필요하다

- data의 최대, 최소가 다름 -> scale에 따른 y 값의 영향
- ordinal, nominal한 값들을 어떻게 표현할 것인가?
- 잘못 기입된 값처리
- 값이 없는 경우는 어떻게 처리해야하나
- 극단적으로 크고 작은값은 어떻게 처리하나

## Coding: the way to bring data by using Pandas

```python
import pandas as pd
```

```python
# data 주소를 변수에 할당해준다
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
# pandas로 데이터를 읽어오는데 pd.read_csv를 사용 ()안에는 1. 읽어올 데이터 주소, 2. seperation을 어떻게 할
# 것인지, 3. header의 유무
# \s+는 regularexpression이다 \s -> 공백, + -> 여러게 있으면 분할한다는 뜻
df_data = pd.read_csv(data_url, sep='\s+', header= None)
```

```python
# data의 첫 5줄을 일어오는 방법
df_data.head()
```

```python
# .columns로 header 부분을 추가해줄수 있다
df_data.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 
                   'TAX', 'PTRATO', 'B', 'LSTAT', 'MEDV']
```

```python
# 확인해보자 head 안에 index를 넣어주면 해당 인덱스 만큼의 data row를 보여준다
df_data.head(20)
```
