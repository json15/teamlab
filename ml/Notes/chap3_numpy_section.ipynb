## What is Numpy?

#### Numpy 의 특징
- 일반 list에 비해 빠르고, 메모리 효율적
- 분복문 없이 데이터 배열에 대한 처리 지원

```python
import numpy as np
# 이렇게 str 타입이 들어가도 자동으로 전부 float으로 바뀐다
test_array = np.array(['1', '4', 5, 8], float)# float자리에 int, str을 써도 바뀐다
print(test_array)
type(test_array[3])
```

#### Numpy Array와 Python list의 메모리 사용법

- data 끼리 다 붙어 있기 때문에, type이 다동일 해야하지만 속도가 빠르다
Python List

#### Array의 dtype
- Ndarray의 single element가 가지는 type
- 각 element가 차지하는 memory의 크기가 결정된다!!
    - 큰수를 쓸것인지 작은수를 쓸것인지에 따라 data type을 조정해주는 것이 좋다!
- c 언어의 타입과 compatible(호환이 가능하다)



```python
# nbytes로 ndarray object의 memory 크기를 반환할수 있다
print(np.array([[1,2,3],[1,2,3],[1,2,3]], dtype=np.float32).nbytes)
print(np.array([[1,2,3],[1,2,3],[1,2,3]], dtype=np.int8).nbytes)
print(np.array([[1,2,3],[1,2,3],[1,2,3]], dtype=np.float64).nbytes)
```
