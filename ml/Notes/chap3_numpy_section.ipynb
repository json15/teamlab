## What is Numpy?

#### Numpy 의 특징
- 일반 list에 비해 빠르고, 메모리 효율적
- 분복문 없이 데이터 배열에 대한 처리 지원
    - for문, loop이 없당!
- 선형대수와 관련된 다양한 기능 제공
- c, c++, 포트란 등의 언어와 통합 가능

## NdArray

#### Import 하는 법
import numpy as np

#### NdArray 의 특징
- np.array 함수를 활용해서 배열을 생성한다
- 하나의 data type만 넣을수 있다
    - 속도를 위해서!
- Dynamic typing not supported!
    - C Array를 사용하여 배열을 생성하기 때문이다


```python
import numpy as np
# 이렇게 str 타입이 들어가도 자동으로 전부 float으로 바뀐다
test_array = np.array(['1', '4', 5, 8], float)# float자리에 int, str을 써도 바뀐다
print(test_array)
type(test_array[3])
```

```python
test_array.dtype
#전체 array의 type 반환하는 법!
```

#### Numpy Array와 Python list의 메모리 사용법
Numpy Array
- Data가 붙어있다
- data 끼리 다 붙어 있기 때문에, type이 다동일 해야하지만 속도가 빠르다
Python List
- 밖에 있는 데이터를 참조하는 방식이다
- 그렇기 때문에 여러가지 type이 다들어갈수 있는것이다, 단지 일일이 찾아와야하기때문에 속도가 헬
